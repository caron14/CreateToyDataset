```python
"""
    Import Library
"""
##-- Pandas
import pandas as pd
##-- NumPy
import numpy as np
##-- Matplotlib
import matplotlib.pylab as plt
import seaborn as sns
##-- Scikit-learn
import sklearn
```

```python
def func(param, X):
    """
        y = w0 + w1*x + w2*x^2 + w3*exp(x)
    """
    return param[0] + param[1]*X + param[2]*np.power(X, 2) + param[3]*np.exp(X)
```

```python
##-- Set Random Seed
np.random.seed(seed=99)

x = np.arange(0, 1, 0.01)
param = [-1.0, 1.0, 2.0, 3.0]

"""
    (x, y_model):  dataset of the Model Function
    (x, y_train):  dataset for training, "y_model + random noise"
"""
y_model = func(param, x)
y_train = func(param, x) + np.random.normal(loc=0, scale=1.0, size=len(x))
```

```python
plt.figure(figsize=(5, 5), dpi=100)
sns.set()
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-1.5, 12.5)
plt.scatter(x, y_train, lw=1, color="b", label="training dataset")
plt.plot(x, y_model, lw=3, color="r", label="model function")
plt.savefig("toy-dataset.png")
plt.legend()
plt.show()
```
